## **1-e. 범주형 변수에 대해 Label Encoding을 수행해 주세요.**

````python
from sklearn.preprocessing import LabelEncoder
````

````python
# 범주형 변수 리스트
cat = ['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety', 'class']

# LabelEncoder 객체 생성
encoder=LabelEncoder()

# 각 범주형 변수에 대해 레이블 인코딩 수행
for i in cat:
  encoder.fit(df[i])
  df[i]=encoder.transform(df[i])
````

# **3. 앙상블**

## **3-a. breast_cancer 데이터셋을 ```load()``` 함수를 사용해서 불러온 후, 훈련 세트와 테스트 세트로 나눠주세요.**
(test_size = 0.2, random_state = 42)




````python
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
````

````python
cancer = load_breast_cancer()

X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, test_size=0.2, random_state=42)
````

## **3-b. 로지스틱 회귀, KNN, 결정 트리를 개별 모델로 사용하여 개별 모델의 학습/예측/평가를 진행해 보세요.**
(평가는 정확도가 출력되도록 해주세요.)
- 로지스틱 회귀: solver = "liblinear"
- KNN: n_neighbors = 16
- 결정 트리: random_state = 12

````python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
````

````python
# 개별 모델
lr_clf=LogisticRegression(solver="liblinear")
knn_clf=KNeighborsClassifier(n_neighbors=16)
dt_clf=DecisionTreeClassifier(random_state = 12)

# 학습
lr_clf.fit(X_train, y_train)
knn_clf.fit(X_train, y_train)
dt_clf.fit(X_train, y_train)
# 예측
pred_lr=lr_clf.predict(X_test)
pred_knn=knn_clf.predict(X_test)
pred_dt=dt_clf.predict(X_test)
# 평가
print('로지스틱 회귀 정확도: {}'.format(accuracy_score(y_test,pred_lr)))
print('KNN 정확도: {}'.format(accuracy_score(y_test,pred_knn)))
print('결정 트리 정확도: {}'.format(accuracy_score(y_test,pred_dt)))
````

## **3-c. 로지스틱 회귀와 결정 트리를 기반으로 하여 '하드 보팅' 방식으로 보팅 분류기를 생성한 후, 학습/예측/평가를 진행해 보세요.**
(평가는 정확도가 출력되도록 해주세요.)

````python
from sklearn.ensemble import VotingClassifier
# 분류기 생성
h_clf=VotingClassifier(estimators=[('LR',lr_clf),('DT',dt_clf)],voting='hard')

# 학습/예측/평가
h_clf.fit(X_train, y_train)
pred_h=h_clf.predict(X_test)
print('하드 보팅 정확도: {}'.format(accuracy_score(y_test,pred_h)))
````

## **3-d. KNN과 결정트리를 기반으로 하여 '소프트 보팅' 방식으로 보팅 분류기를 생성한 후, 학습/예측/평가를 진행해보세요.**
(평가는 정확도가 출력되도록 해주세요.)

````python
from sklearn.ensemble import VotingClassifier
# 분류기 생성
s_clf=VotingClassifier(estimators=[('KNN',knn_clf),('DT',dt_clf)],voting='soft')

# 학습/예측/평가
s_clf.fit(X_train, y_train)
pred_s=s_clf.predict(X_test)
print('소프트 보팅 정확도: {}'.format(accuracy_score(y_test,pred_s)))
````
